```python
import fitz  # PyMuPDF
import re

def extract_complete_headnotes(pdf_file_path):
    with fitz.open(pdf_file_path) as doc:
        full_text = ""
        for page in doc:
            full_text += page.get_text()

    # 修正案例分割逻辑，确保能够正确分割所有案例
    cases = re.split(r'(?:Core Terms|End of Document)', full_text)[1:]  # 忽略首部非案例文本

    case_count = 0

    for case_content in cases:
        # 使用更广泛的正则表达式匹配 "Headnotes" 内容，终止条件改为 "Syllabus"
        headnotes_matches = re.findall(r'(?<=Headnotes\n)([\s\S]*?)(?=\nSyllabus|$)', case_content)
        
        if headnotes_matches:
            case_count += 1
            print(f"Case {case_count}:")

            for headnotes_content in headnotes_matches:
                # 提取包含 ">" 的行，同时保留 ">" 前的文本
                lines_with_arrow = re.findall(r'(.*?>\s*.+)', headnotes_content)
                
                for line in lines_with_arrow:
                    print(line.strip())
            
            print("--------------------------------------------------")

    print(f"Total cases processed: {case_count}")

# 文件路径
pdf_file_path = r"C:\Users\User\Dropbox\textmining1\PDF\Files (100).PDF"

extract_complete_headnotes(pdf_file_path)
```

```python
import fitz  # PyMuPDF
import re

def extract_cases_and_check_content(pdf_file_path):
    with fitz.open(pdf_file_path) as doc:
        full_text = ""
        for page in doc:
            full_text += page.get_text()

    # 分割案例
    cases = re.split(r'(?:Core Terms|End of Document)', full_text)

    case_number = 0
    for i in range(1, len(cases), 2):  # 跳过"Core Terms"本身，直接处理每个案例的内容
        case_content = cases[i].strip()
        if case_content:  # 检查案例内容是否为空
            case_number += 1
            if '>' in case_content:
                print(f"Case {case_number}: Contains relevant content\n")
                lines_with_arrow = re.findall(r'(.*?>\s*.+)', case_content)
                for line in lines_with_arrow:
                    print(line.strip())
            else:
                print(f"Case {case_number}: Empty Case\n")
            print("--------------------------------------------------")
        else:
            # 如果案例内容为空（可能是文件的最后），则不增加案例编号
            continue

    print(f"Total cases processed: {case_number}")

# 文件路径
pdf_file_path = r"C:\Users\User\Dropbox\textmining1\PDF\Files (100).PDF"

extract_cases_and_check_content(pdf_file_path)
```
